```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Plot show picture colors
dir = 'data/train/dog/dog.10.jpg'
img = cv2.imread(dir)
plt.figure(figsize=(3,3))
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB));
```

```python
#Plot grayscale
imgGray = cv2.imread(dir, cv2.IMREAD_GRAYSCALE)
plt.figure(figsize=(3,3))
plt.imshow(imgGray, cmap='gray');
```

```python
# Crop image
START_Y = 50
END_Y = 150
START_X = 50
END_X = 150
img = cv2.imread(dir)
crop = img[START_Y:END_Y, START_X:END_X,:]
plt.figure(figsize=(3,3))
plt.imshow(cv2.cvtColor(crop, cv2.COLOR_BGR2RGB)), plt.title('Crop image');
```

```python
#Placee crop to original
imgModify = img.copy()
imgModify[:END_Y-START_Y, :END_X-START_X, :] = crop
plt.figure(figsize=(3,3))
plt.imshow(cv2.cvtColor(imgModify, cv2.COLOR_BGR2RGB)),plt.title('Modify');
```

```python
#Channel
img = cv2.imread(dir)
b = img[...,0]
g = img[...,1]
r = img[...,2]
plt.figure(figsize=(12,8))
plt.subplot(1,4,1),plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB)),plt.title('Original')
plt.subplot(1,4,2),plt.imshow(b, cmap='gray'),plt.title('b')
plt.subplot(1,4,3),plt.imshow(g, cmap='gray'),plt.title('g')
plt.subplot(1,4,4),plt.imshow(r, cmap='gray'),plt.title('r');
```

```python
#Add more red 
imgR = img.copy()
imgR[...,2] = 255
plt.figure(figsize=(5,2))
plt.subplot(1,2,1),plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB)),plt.title('Original')
plt.subplot(1,2,2),plt.imshow(cv2.cvtColor(imgR, cv2.COLOR_BGR2RGB)),plt.title('More Red');
```
